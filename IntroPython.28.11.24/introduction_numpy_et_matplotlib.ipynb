# Importer les packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Tra√ßer une fonction sinus dans un graphique

```python
# On d√©finit le domaine (l'abscisse)
x = np.linspace(0, 6*np.pi, 1000)
#x = np.arange(0, 2*np.pi, 0.01)

# On d√©finit les ordonn√©es
y = np.sin(x)
f = np.cos(x)

# Change the style of the plot to dark background
plt.style.use('dark_background')

plt.plot(x, y, color = '#E1439E', linewidth = 2, linestyle = '--', label = 'sin(x)')
plt.plot(x, f, color = 'green', linewidth = 2, linestyle = '-', label = 'cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fonction sinus', fontsize = 20, color = 'blue', style = 'italic')
plt.grid()
plt.legend()
plt.show()
```
